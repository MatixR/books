```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0, 2*np.pi, 90)
```

```python
tau = np.radians(15)
k = (2 * np.pi - tau) / np.pi 
e = 0.025
l = 1
a = 2 * l * np.power(e + 1, k - 1) / np.power(2, k)
```

```python
circle = a * np.cos(theta) + a * 1j * np.sin(theta)
```

```python
airfoil = np.power(circle - a, k) / np.power(circle - a * e, k - 1) + l
```

```python
plt.plot(circle.real, circle.imag, color='r')
plt.plot(airfoil.real, airfoil.imag, color='b')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axis('equal')
plt.show()
```
